### a) Sketching on the Complex Plane

To sketch these on the complex plane, the x-axis represents the real part, and the y-axis represents the imaginary part. As a result:  

- $z_1 = 1 + i$ can be represented at the coordinate $(1, 1)$.  
- $z_2 = -\sqrt{3} + 2i$ at $(-\sqrt{3}, 2)$.  
- $z_3 = -3i$ at $(0, -3)$.  

To find the polar representation of these numbers, we need to express them in terms of angle $\theta$ and its distance from the origin $r$:  

$$
r(\cos(\theta) + i\sin(\theta)) = r\cos(\theta) + r\sin(\theta)i = a + bi
$$

Thus, we have:  

- $r_1 = |z_1| = \sqrt{1^2 + 1^2} = \sqrt{2}$  
  $$
  \theta = \arg(z_1) = \tan^{-1}\left(\frac{1}{1}\right) \approx 0.7854 \text{ rad}
  $$
  $$
  z_1 \approx \sqrt{2} e^{i 0.79}
  $$

- $r_2 = |z_2| = \sqrt{3+4} = \sqrt{7}$  
  $$
  \theta = \arg(z_2) = \tan^{-1}\left(\frac{2}{-\sqrt{3}}\right) \approx -0.8571 \text{ rad}
  $$
  $$
  z_2 \approx \sqrt{7} e^{i (-0.86)}
  $$

- $r_3 = |z_3| = \sqrt{9} = 3$  
  $$
  \theta = \arg(z_3) = \tan^{-1}\left(\frac{-3}{0}\right)
  $$
  Since this is division by zero, we need to consider:  
  - If $b < 0$, then $\arg(z) = -\frac{\pi}{2}$ or $\frac{3\pi}{2}$.  

  $$
  z_3 \approx 3 e^{i (-\pi/2)}
  $$

### b) Verification of Multiplication

#### Direct Computation

$$
z_1z_2 = (1 + i)(-\sqrt{3} + 2i)
$$

Expanding:

$$
z_1z_2 = -\sqrt{3} + 2i - i\sqrt{3} + 2i^2
$$

Since $i^2 = -1$, we get:

$$
z_1z_2 = -\sqrt{3} - 2 + i(2 - \sqrt{3})
$$

Magnitude:

$$
|z_1z_2| = \sqrt{14}
$$

Angle:

$$
\theta = \arg(z_1z_2) = \tan^{-1} \left(\frac{2 - \sqrt{3}}{-\sqrt{3} - 2}\right) \approx -0.07 \text{ rad}
$$

Thus,

$$
z_1z_2 \approx \sqrt{14} e^{-i 0.07} \quad (1)
$$

#### Polar Form Computation

$$
z_1 \approx \sqrt{2} e^{i 0.79}, \quad z_2 \approx \sqrt{7} e^{i (-0.86)}
$$

Multiplying in polar form:

$$
z_1z_2 = \sqrt{2} e^{i 0.79} \cdot \sqrt{7} e^{i (-0.86)}
$$

$$
z_1z_2 = \sqrt{14} e^{i (0.79 - 0.86)}
$$

$$
z_1z_2 = \sqrt{14} e^{-0.07 i} \quad (2)
$$

From (1) and (2), we can conclude that the two expressions are equal!

### Exercise 3. Show that $ae^{ix} + be^{-ix} = c\cos(x) + d\sin(x)$, where $c, d \in C$. What are the values of c, d in terms of a, b?

We have shown earlier that:

$$
e^{ix} = \cos(x) + i\sin(x)
$$

$$
e^{-ix} = \cos(x) - i\sin(x)
$$

Expanding:

$$
a e^{ix} = a (\cos(x) + i\sin(x)) = a\cos(x) + a i \sin(x)
$$

$$
b e^{-ix} = b (\cos(x) - i\sin(x)) = b\cos(x) - b i \sin(x)
$$

Adding both expressions:

$$
a e^{ix} + b e^{-ix} = a\cos(x) + a i \sin(x) + b\cos(x) - b i \sin(x)
$$

$$
= (a+b) \cos(x) + (a i - b i) \sin(x) \quad (*)
$$

Let $c = a + b$ and $d = a i - b i$, we can substitute these into (*), giving:

$$
a e^{ix} + b e^{-ix} = c \cos(x) + d \sin(x)
$$
